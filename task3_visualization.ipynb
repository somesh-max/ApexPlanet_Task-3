```python
# Import required libraries

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
plt.style.use("ggplot")
sns.set_theme(style="whitegrid")
```

```python
# Load the dataset

df = pd.read_csv(r"C:\Users\somes\OneDrive\Pictures\Attachments\Desktop\Documents\apex planet\task-3\SuperMarket Analysis.csv")

# Display first five rows

df.head()
```

```python
# Display dataset information

df.info()

# Display statistical summary

df.describe()
```

```python
# Check missing values

df.isnull().sum()
```

```python
# Histogram
# Visualize the distribution of Sales.

plt.figure(figsize=(8,5))

plt.hist(df["Sales"], bins=20)

plt.title("Distribution of Sales")

plt.xlabel("Sales")

plt.ylabel("Frequency")

plt.show()
```

```python
# Bar Chart
# Compare total sales across different cities.

plt.figure(figsize=(8,5))

sns.barplot(
    data=df,
    x="City",
    y="Sales",
    estimator=sum
)

plt.title("Total Sales by City")

plt.xlabel("City")

plt.ylabel("Sales")

plt.show()
```

```python
# Pie Chart
# Show the percentage of each payment method.

plt.figure(figsize=(6,6))

df["Payment"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%"
)

plt.title("Payment Method Distribution")

plt.ylabel("")

plt.show()
```

```python
# Product Line Analysis
# Compare sales across different product lines.

plt.figure(figsize=(10,5))

sns.barplot(
    data=df,
    x="Product line",
    y="Sales",
    estimator=sum
)

plt.xticks(rotation=30)

plt.title("Sales by Product Line")

plt.xlabel("Product Line")

plt.ylabel("Sales")

plt.show()
```

```python
# Rating Distribution
# Display the distribution of customer ratings.

plt.figure(figsize=(8,5))

sns.histplot(df["Rating"], bins=10)

plt.title("Customer Rating Distribution")

plt.xlabel("Rating")

plt.ylabel("Frequency")

plt.show()
```

```python
# Scatter Plot
# Analyze the relationship between Quantity and Sales.

plt.figure(figsize=(8,5))

plt.scatter(df["Quantity"], df["Sales"])

plt.title("Quantity vs Sales")

plt.xlabel("Quantity")

plt.ylabel("Sales")

plt.show()
```

```python
# Box Plot
# Detect outliers in the Sales column.

plt.figure(figsize=(6,5))

sns.boxplot(y=df["Sales"])

plt.title("Box Plot of Sales")

plt.show()
```

```python
# Correlation Heatmap
# Display correlation among numerical variables.

plt.figure(figsize=(10,8))

sns.heatmap(
    df.select_dtypes(include=np.number).corr(),
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Heatmap")

plt.show()
```

```python
# Gender Distribution
# Count male and female customers.

plt.figure(figsize=(6,5))

sns.countplot(
    data=df,
    x="Gender"
)

plt.title("Customer Distribution by Gender")

plt.show()
```

```python
# Branch-wise Sales
# Compare sales across supermarket branches.

plt.figure(figsize=(6,5))

sns.barplot(
    data=df,
    x="Branch",
    y="Sales",
    estimator=sum
)

plt.title("Branch-wise Sales")

plt.show()
```
